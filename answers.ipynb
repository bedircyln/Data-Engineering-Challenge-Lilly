# FIRST TASK

Create a query that calculates the average number of goals per game between 1900 and 2000.



```python
# Converting 'date' columns datas from string to date format
results_df['date'] = pd.to_datetime(results_df['date'])

# Filtering the 'results' dataframe to see the results between 1900-2000
results_1900to2000 = results_df[(results_df['date'] >= '1900-01-01') & (results_df['date'] < '2001-01-01')]
#results_1900to2000.head()

# Counting the number of the games between 1900-2000
number_of_games_1900to2000 = results_1900to2000.shape[0]
print('Number of the games between 1900 and 2000:',number_of_games_1900to2000)

# Calculating the total number of the goals scored between 1900 and 2000
total_goals_1900to2000 = results_1900to2000['home_score'].sum() + results_1900to2000['away_score'].sum()

# Calculating the average number of goals per game between 1900 and 2000
average_goals_1900to2000 = total_goals_1900to2000 / number_of_games_1900to2000
print("Average number of goals per game between 1900 and 2000 is =>",average_goals_1900to2000)

# Transforming the "date" type back to its string format
results_df['date'] = results_df['date'].dt.strftime('%Y-%m-%d')
```


# FIFTH TASK



Create a query that identifies the top goalscorer by tournament, and what percentage that equates to for all goals scored in the tournament.



1) Calculate the total number of the goals by tournament.

```python
# Calculating the number of the goals for each game
# And adding a new column "total_goals" to the results dataframe
results_df["total_goals"] = results_df["home_score"] + results_df["away_score"]

# Grouping the data by 'tournament' column and summing up the total goals for each tournament
# after the grouping, tournament becomes an index so we need to reset it with reset_index() and make it a new column named 'total_goals'
tournament_goals = results_df.groupby('tournament')['total_goals'].sum().reset_index(name='total_goals')
#tournament_goals
```

2) Merge the goalscorers with results data so we can see the relationships between the goalscorer player and game results 

```python
tournaments_and_goalscorers_df = goalscorers_df.merge(results_df, on='Game_ID', how='left')
```

3) Calculate the number of the goals for each player by tournament

```python
# Grouping the data by 'tournament' and 'scorer'
# This groups the data based on each unique combination of tournament and scorer.
player_and_tournament = tournaments_and_goalscorers_df.groupby(['tournament','scorer']).size().reset_index(name='goal_count')

player_and_tournament
```

4) Calculate the maximum number of goals scored by one player for each tournament.

```python
max_goals_tournament = player_and_tournament.groupby('tournament')['goal_count'].max().reset_index(name='max_goals')
max_goals_tournament
```

5) Find which players are the top scorers of each tournament

```python
# Merge the top scorers of each tournament with the maximum goals for that tournament
# 'player_and_tournament' contains players and their total goals for each tournament
# 'max_goals_tournament' contains the maximum number of goals scored in each tournament
# Do a left join so we can the 'max_goals' column for each tournament 
top_scorers_tournament = player_and_tournament.merge(max_goals_tournament, on='tournament', how='left')

# To find the top scorer of each tournament, 
# we can filter the dataframe to keep only rows where the player's 'goal_count' equals the 'max_goals' for that tournament
best_scorer_tournament = top_scorers_tournament[(top_scorers_tournament['goal_count']) == (top_scorers_tournament['max_goals'])]
best_scorer_tournament
```

6) Now we have the top scorers ("best_scorer_tournament") and the total number of goals ("tournament_goals") for each tournament in separate sets. We can merge the tournament goals with best_scorer to do the future calculations on one dataframe
    


```python
best_scorer_tournament = best_scorer_tournament.merge(tournament_goals, on='tournament' )
best_scorer_tournament
```

7) Calculate the percentage that shows each top scorer's contribution to the total goals in the tournament (convert in percentage). And add the goal percentage in dataframe as column

```python
best_scorer_tournament['goal_percentage'] = (best_scorer_tournament['goal_count'] / best_scorer_tournament['total_goals']) * 100
best_scorer_tournament
```

Create and additional column that flags records with data quality issues

Resolve the identified quality issues

I didn't want to change the order of the tasks so I've put the data quality issues and solutions after the 5th question. This can be done before doing the calculations for the questions (I can also move the cells).

Before doing any operations, I have checked some potential data issues in each dataframe (I wrote the infos for each datasets before the first tasks solution, so I didn't need to write codes for them)
1) home_score and away_score in results dataframe are never negative numbers and the highest goals scored in one game is with 31 home goals which actually happened in Australia - American Samoa game. So there aren't any problems in score datas.
2)  minute in goalscorers dataframe starts from '1' (which is the earliest possible minute that a player can score in football) and ends with '122' so there aren't any problematic numbers. 
3) There are some missing 'minute', 'own_goal' and 'penalty' values in goalscorers datas.
4)  There are some missing 'first_shooter' values in shootouts datas.

1) Finding the duplicated datas

```python
# Find the duplicated rows in 'results' dataframe and flag them
duplicates_results = results_df.duplicated(subset=['Game_ID'], keep=False)
results_df['data_quality_issues'] = duplicates_results
results_df[results_df['data_quality_issues'] == True]
```

```python
# Fixing the duplicated rows in results dataframe
results_df = results_df.drop_duplicates(subset=['Game_ID'], keep='first')

# Unflag the duplicated rows after dropping the one of the copy's  
duplicates_results = results_df.duplicated(subset=['Game_ID'], keep=False)
results_df['data_quality_issues'] = duplicates_results

results_df[results_df['data_quality_issues'] == True]
```

```python
# Find the duplicated rows in shootouts and flag them 
duplicates_shootouts = shootout_df.duplicated(subset=['Game_ID','first_shooter'], keep=False)
shootout_df['data_quality_issues'] = duplicates_shootouts
shootout_df[shootout_df['data_quality_issues'] == True]
```

```python
# Find the duplicated rows in goalscorers and flag them 
duplicates_goalscorers = goalscorers_df.duplicated(subset=['Game_ID', 'minute', 'scorer'], keep=False)
goalscorers_df['data_quality_issues'] = duplicates_goalscorers
# In goalscorers dataset, there are some 'NaN' minute values (which is another quality issue), 
# One player can score more than one goal in the same game but cannot score more than two goals in one game in same minutes.
# So I've filtered NaN minutes in dataset so it only shows the players who scored in same minute in the same game
goalscorers_df[(goalscorers_df['data_quality_issues'] == True) & (goalscorers_df['minute'].notna())]
```
